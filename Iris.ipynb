```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x="SepalLengthCm",bins=15,kde=True)
plt.title("Distribution of SepalLengthCm")
plt.xlabel("SepalLengthCm")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x="Species",bins=15,kde=True,color="orange")
plt.title("Distribution of Species")
plt.xlabel("Species")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
continuous_columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
for column in continuous_columns:
 plt.figure(figsize=(10,5))
 
 sns.histplot(data=df,x=column,bins=15,kde=True,color='brown')
 plt.title(f"Histogram of {column}")
 plt.xlabel(column)
 plt.ylabel('Frequency') 
 plt.grid(True)
 plt.show()
 
 plt.figure(figsize=(8,4))
 sns.boxplot(data=df,y=column)
 plt.title(f"Box Plot of {column}")
 plt.xlabel(column) 
 plt.grid(True)
 plt.show()
```
